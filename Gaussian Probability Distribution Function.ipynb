```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Gaussian Probability Distribution Function (PDF)

```python
x = np.linspace(-5, 5)

def PDF(mu, sd, n):
    #mean : mu
    #standard diviation : sd
    
    y = 1 / ( sd * np.sqrt(2 * np.pi  ) ) 
    y = y * np.exp(  - (n - mu)**2 /  (2 * sd**2) )
                        
    return y
```

```python
mu = 0
sd = 0.2
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)

plt.plot(x, PDF(0, 0.5, x), 'r-', label="$\mu = 0, \sigma = 0.5$")
plt.plot(x, PDF(0, 1.0, x), 'g-', label="$\mu = 0, \sigma = 1.0$")
plt.plot(x, PDF(0, 1.5, x), 'b-', label="$\mu = 0, \sigma = 1.5$")
plt.plot(x, PDF(-2, 1.0, x), 'c-', label="$\mu =-2, \sigma = 1.5$")
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gaussian Probability Distribution Function (PDF)')

```
